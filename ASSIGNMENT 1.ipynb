Generate the series of dates from 1st Feb, 2023 to 1st March, 2023 (both inclusive)

```python
import datetime
import pandas as pd
 

date = datetime.datetime.strptime("01-2-2023", "%d-%m-%Y")
 
 
date_generated = pd.date_range(date, periods=29)
print(date_generated.strftime("%d-%m-%Y"))
```
